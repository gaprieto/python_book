## Cargar datos una sola vez

```python
import numpy as np

# Nombres de archivos
tfile = '../data/MON_T_CRU_19012015.csv'
pfile = '../data/MON_P_CRU_19012015.csv'
sta_file = '../data/BASIN_CHARACTERISTICS.csv'


# Read station file
basin = np.loadtxt(sta_file,skiprows=1,delimiter=',')
sta  = basin[:,0]
lon  = basin[:,1]
lat  = basin[:,2]
area = basin[:,3]
elev = basin[:,4]

# Read data file
Temp = np.loadtxt(tfile, skiprows = 1, delimiter = ',')
Prec = np.loadtxt(pfile, skiprows = 1, delimiter = ',')
basin = np.loadtxt(sta_file, skiprows = 1, delimiter = ',')

# Datos
T = Temp[:,1:]
P = Prec[:,1:]

nsta,npts = np.shape(T)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.io.img_tiles as tiles

# Select station to plot 
ista = 801

# Create map
fig = plt.figure()
ax  = fig.add_subplot(1,1,1,projection=ccrs.PlateCarree())
ax.set_extent([-180, 180, -90, 90])
ax.coastlines()
ax.plot(lon,lat,'k^')
ax.plot(lon[ista],lat[ista],'r^')

tiler = tiles.Stamen('terrain-background')

fig = plt.figure()
ax  = fig.add_subplot(1,1,1,projection=ccrs.PlateCarree())
ax.set_extent([lon[ista]-15, lon[ista]+15, lat[ista]-15, lat[ista]+15])
ax.set_title('Near station')
ax.add_image(tiler, 7)
ax.coastlines(color='black',resolution='10m')
ax.plot(lon,lat,'k^')
ax.plot(lon[ista],lat[ista],'r^')

plt.show()
```

```python
import pygmt
import numpy as np


reg = [lon[ista]-15, lon[ista]+15, lat[ista]-15, lat[ista]+15]

fig = pygmt.Figure()
fig.coast(
    region='g',
    projection='R0/8i',
    shorelines=True,
    water='lightblue',
    frame=True,
)


fig.plot(lon,lat,
    style='t0.1i',
    color='black',
        )
fig.plot(lon[ista],lat[ista],
    style='t0.4i',
    color='red',
        )

fig.shift_origin(xshift='2.i',yshift='-3.5i')  # Shift for next call

fig.grdimage(
    '@earth_relief_05m',
    region=reg,
    projection='M4i',
    cmap='topo',
    shading=True,
)
fig.coast(
    region=reg,
    projection='M4i',
    shorelines=True,
    water='lightblue',
    frame='f'
)

fig.plot(lon,lat,
    style='t0.2i',
    color='black',
        )
fig.plot(lon[ista],lat[ista],
    style='t0.4i',
    color='red',
        )


fig.show()
```

```python
# ahora graficamos los datos de la estación
# 115 años corresponden a 1380 datos

t = np.linspace(1901,2016,npts,endpoint=False)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(t, T[ista,:],color='r')
ax.set_ylabel('Temperatura', color='r')

ax.tick_params(axis='y', labelcolor='r')

ax2 = ax.twinx()  
ax2.plot(t, P[ista,:],color='b')
ax2.set_ylabel('Precipitación', color='b')
ax2.tick_params(axis='y', labelcolor='b')

#fig.savefig('tp')

# limitamos el eje x para obtener datos de una época deseada
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(t, T[ista,:],color='r')
ax.set_ylabel('Temperatura', color='r')

ax.tick_params(axis='y', labelcolor='r')

ax2 = ax.twinx()  
ax2.plot(t, P[ista,:],color='b')
ax2.set_ylabel('Precipitación', color='b')
ax2.tick_params(axis='y', labelcolor='b')

ax.set_xlim(1980,2016)
#fig.savefig('tpzoom')
```
